#### Problem 1: Find the total number of crimes recorded in the CRIME table.

```python
%sql select count(distinct CASE_NUMBER) as 'total_number_of_crimes' from CHICAGO_CRIME_DATA
```

#### Problem 2: List community areas with per capita income less than 11000.

```python
%sql select COMMUNITY_AREA_NAME from CENSUS_DATA where PER_CAPITA_INCOME<11000
```

#### Problem 3: List all case numbers for crimes involving minors?

```python
%sql select PRIMARY_TYPE from CHICAGO_CRIME_DATA group by(PRIMARY_TYPE)
```

```python
%sql select CASE_NUMBER from CHICAGO_CRIME_DATA where DESCRIPTION like '%MINOR%'
```

#### Problem 4: List all kidnapping crimes involving a child?(children are not considered minors for the purposes of crime analysis)

```python
%sql select PRIMARY_TYPE,DESCRIPTION from CHICAGO_CRIME_DATA GROUP BY(PRIMARY_TYPE)
```

```python
%sql select CASE_NUMBER,PRIMARY_TYPE,DESCRIPTION from CHICAGO_CRIME_DATA where PRIMARY_TYPE like'%KIDNAPPING%'
```

#### Problem 5: What kind of crimes were recorded at schools?

```python
%%sql select PRIMARY_TYPE,LOCATION_DESCRIPTION from CHICAGO_CRIME_DATA GROUP BY(LOCATION_DESCRIPTION) 
ORDER BY LOCATION_DESCRIPTION DESC
```

```python
%%sql select CASE_NUMBER,PRIMARY_TYPE,LOCATION_DESCRIPTION from 
CHICAGO_CRIME_DATA where LOCATION_DESCRIPTION like'%SCHOOL%' group by(PRIMARY_TYPE)
```

#### Problem 6: List the average safety score for all types of schools.

```python
%%sql select Elementary_Middle_or_High_School,count(Elementary_Middle_or_High_School) 
from CHICAGO_PUBLIC_SCHOOLS group by(Elementary_Middle_or_High_School)
```

```python
%%sql select Elementary_Middle_or_High_School as School_Types,round(avg(SAFETY_SCORE),2) as Average_safety_score
from CHICAGO_PUBLIC_SCHOOLS group by(Elementary_Middle_or_High_School)
```

#### Problem 7: List 5 community areas with highest % of households below poverty line.

```python
%%sql select COMMUNITY_AREA_NAME, PERCENT_HOUSEHOLDS_BELOW_POVERTY from 
CENSUS_DATA order by PERCENT_HOUSEHOLDS_BELOW_POVERTY desc limit 5
```

#### Problem 8: Which community area(number) is most crime prone?

```python
%%sql select COMMUNITY_AREA_NUMBER ,count(COMMUNITY_AREA_NUMBER) as Number_of_crimes from 
CHICAGO_CRIME_DATA group by(COMMUNITY_AREA_NUMBER) 
order by Number_of_crimes desc
limit 1
```

#### Problem 9: Use a sub-query to find the name of the community area with highest hardship index

```python
%sql select COMMUNITY_AREA_NAME from CENSUS_DATA where HARDSHIP_INDEX=(select max(HARDSHIP_INDEX) from CENSUS_DATA)
```

#### Problem 10: Use a sub-query to determine the Community Area Name with most number of crimes?

```python
%%sql select COMMUNITY_AREA_NAME,ccd.COMMUNITY_AREA_NUMBER,count(CASE_NUMBER) as number_of_crimes
from CENSUS_DATA cd 
join CHICAGO_CRIME_DATA ccd
on cd.COMMUNITY_AREA_NUMBER=ccd.COMMUNITY_AREA_NUMBER 
group by(COMMUNITY_AREA_NAME)
order by number_of_crimes desc
limit 1
```
